```python
import pandas as pd
```

```python
url='https://raw.githubusercontent.com/usama6naseer/largest_object/master/log.csv'
```

```python
def checkCSV(url):
    if url.endswith('.csv'):
        d = pd.read_csv(url)
        return d
    else:
        raise Exception('Invalid format')
        
```

```python
df=checkCSV(url)
```
